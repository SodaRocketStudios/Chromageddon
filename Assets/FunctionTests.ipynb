```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constant multiplication
initialValue = 5
multiplier = 1.025

y = [initialValue]

for i in range(1, 200):
    y.append(y[i-1]*multiplier)

plt.plot(y)
```
